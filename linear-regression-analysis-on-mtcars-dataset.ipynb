```r
data(mtcars)
head(mtcars)
summary(mtcars)
plot(mtcars$hp, mtcars$mpg,
    main = "MPG vs Horsepower",
    xlab = "Horsepower",
    ylab = "Miles Per Gallon",
    col = "blue",
    pch = 19)
model <- lm (mpg ~ hp, data = mtcars)
abline(model, col = "red")
summary(model)
```
